```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../DB/orderHistory.csv')

sales_report = df.groupby('Name')['Number_of_Order'].sum().sort_values()

plt.figure(figsize=(10, 6))
sales_report.plot(kind='bar', color='skyblue')
plt.xlabel('Название товара')
plt.ylabel('Количество продаж')
plt.title('График продаж (в порядке возрастания)')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
df['Total'] = df['Price'] * df['Number_of_Order']

sales_report = df.groupby('Name')['Total'].sum().sort_values()

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid") 
sns.barplot(x=sales_report.index, y=sales_report.values, palette="viridis")
plt.xlabel('Название товара', fontsize=12)
plt.ylabel('Общая сумма продажи', fontsize=12)
plt.title('Общая сумма продажи по каждому товару', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)


plt.show()

```
